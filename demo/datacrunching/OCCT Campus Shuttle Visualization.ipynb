```python
%matplotlib inline
import pandas as pd
#import seaborn as sns #nicer plots, auto integrates with Pandas
```

```python
df = pd.read_csv("occt.csv", names=['lat', 'long']) #Isn't this easy? :D
```

```python
df.plot.scatter("lat","long")
```

```python
df = df.loc[ (df.lat > 42.082) & (df.lat < 42.093) ]
df = df.loc[ (df.long > -75.979) & (df.long < -75.962) ]
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors
output = plt.hist2d(df.long, df.lat, bins=100, norm=colors.LogNorm())
```
